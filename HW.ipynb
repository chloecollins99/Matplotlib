```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn


# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_data = pd.read_csv(city_data_to_load)
ride_data = pd.read_csv(ride_data_to_load)

# Combine the data into a single dataset
merge_df = pd.merge(ride_data, city_data,how="left",on =["city", "city"])

# Display the data table for preview
#city_data["city"].value_counts()
#ride_data["city"].value_counts()
#merge_df["city"].value_counts()
merge_df.head()
```

```python
rural = merge_df[merge_df['type'] == 'Rural']
suburban = merge_df[merge_df['type'] == 'Suburban']
urban = merge_df[merge_df['type'] == 'Urban']
urban 
```

## Bubble Plot of Ride Sharing Data

```python
#group by city name
each_city = merge_df.groupby(['type', 'city'])

#variable holding the average fare per city
avg_fare = each_city.mean()['fare']

#varibale holding number of rides per city
ride_count = each_city['ride_id'].count()

#variable holding number of drivers per city
driver_count = each_city.mean()['driver_count']



#set dataframe index using existing columns to hold city type
#city_type = merge_df.set_index('city')['type']

# dataframe from above series
city_info = pd.DataFrame({
    "Number of Rides": ride_count,
    "Average Fare": avg_fare,
    "Number of Drivers": driver_count,
    
})

x_urban = city_info.loc['Urban']['Number of Rides']
s_urban= city_info.loc['Urban']['Number of Drivers']
y_urban = city_info.loc['Urban']['Average Fare']

x_suburban = city_info.loc['Suburban']['Number of Rides']
s_suburban= city_info.loc['Suburban']['Number of Drivers']
y_suburban = city_info.loc['Suburban']['Average Fare']

x_rural = city_info.loc['Rural']['Number of Rides']
s_rural= city_info.loc['Rural']['Number of Drivers']
y_rural = city_info.loc['Rural']['Average Fare']
```

```python
city_info.head()
```

```python
plt.scatter(x_urban, y_urban, c="coral", s=s_urban*10, marker="o", edgecolor="black", alpha=.5, label ="Urban")
plt.scatter(x_suburban, y_suburban, c='skyblue', s=s_suburban*10, marker="o", edgecolor="black", alpha=.5, label ="Subrban")
plt.scatter(x_rural, y_rural, c='yellow',s=s_rural*10, marker="o", edgecolor="black", alpha=.5, label ="Rural")

plt.suptitle('Pyber Ride Sharing Data (2016)')

#print scatter plot
plt.title('Note : Circle size correlates with driver count per city.')
plt.xlabel('Total Number of Rides (Per City)')
plt.ylabel('Average Fare ($)')
plt.grid(True)


lgnd = plt.legend(frameon = True, edgecolor = 'black', title="City Types")
lgnd.legendHandles[0]._sizes = [75]
lgnd.legendHandles[1]._sizes = [75]
lgnd.legendHandles[2]._sizes = [75]
```

```python
# Show plot
plt.show()
```
